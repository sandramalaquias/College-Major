# College Majors 

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

    Rank - Rank by median earnings (the dataset is ordered by this column).
    Major_code - Major code.
    Major - Major description.
    Major_category - Category of major.
    Total - Total number of people with major.
    Sample_size - Sample size (unweighted) of full-time.
    Men - Male graduates.
    Women - Female graduates.
    ShareWomen - Women as share of total.
    Employed - Number employed.
    Median - Median salary of full-time, year-round workers.
    Low_wage_jobs - Number in low-wage service jobs.
    Full_time - Number employed 35 hours or more.
    Part_time - Number employed less than 35 hours.
    
Using visualizations, we can start to explore questions from the dataset like:

    
    Do students in more popular majors make more money?
        Using scatter plots
    How many majors are predominantly male? Predominantly female?
        Using histograms
    Which category of majors have the most students?
        Using bar plots

```python
#Sample Size = Sample size (unweighted) of full-time.
#Median - Median salary of full-time, year-round workers.

recent_grads.plot(x='Total', y='Median', kind='scatter', title='Median vs. Total', figsize=(15, 8))  
```

There´s a cluster of markers creating a vertical line close to total = 0. This is because there are not that many people with these majors. The median there varies quite a bit from 20k USD to 80k USD.

However, we can clearly see there are a lot more majors with high median where the total value is low (closer to 0). On the other hand, majors with the highest values of total (= most popular majors) are pretty low on the median scale.

So we can say that more popular majors are not a warranty of getting a high paying job – quite the opposite.

```python
# Sample Size = Sample size (unweighted) of full-time.
# Unemployment_rate.

recent_grads.plot(x='Total', y='Unemployment_rate', kind='scatter', title='Unemployment Rate x Total', figsize=(15, 8))
```

So we can say that more popular majors are not a warranty of getting a less umployment rate.

Let's using the Histogram graph to answer

    What percent of majors are predominantly male? Predominantly female?
    What's the most common median salary range? (using Median column)


```python
recent_grads["ShareWomen"].hist(bins=5, range=(0, 1))
plt.title('ShareWomen')
plt.xlabel('ShareWomen')
plt.ylabel('Frequency')
```

If we wanted to get a precise number of predominantly female majors, it would be better to filter for rows where the share of women column is greater than 0.5 and then count the rows. However, we can get an estimate just by looking at the chart – it looks like over 90 majors have more than 50% of women and about 83 are predominantly of men.

```python
recent_grads["Median"].hist(bins=6, range=(20000, 80000))

recent_grads["Median"].hist(bins=6, range=(20000, 80000), color='blue')
plt.title('Median Salary')
plt.xlabel('Median')
plt.title('Median Salary')
plt.xlabel('Median')
plt.ylabel('Frequency')
```

According to this plot, the most frequent median salary is 30-40k USD and about 80 majors earn this money.

* Can be a positive linear regression between Total an Employment that is when Total increase then the Employment increase too.
* Can be a negative linear regression between Median and Employment that is when Median increase then Employment decrease.
